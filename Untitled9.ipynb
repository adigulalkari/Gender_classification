```python
from datasets import load_dataset

ds = load_dataset("tuandunghcmut/PETA")
```

```python
import random
from datasets import DatasetDict

# Shuffle the dataset first
shuffled_ds = ds['train'].shuffle(seed=42)  # Shuffle the dataset with a fixed seed for reproducibility

# Select the first 400 entries from the shuffled dataset
sampled_ds = shuffled_ds.select(range(1000))
```

```python
# Function to create the new column
def add_gender_column(example):
    if example['personalMale'] == 1:
        example['gender_male'] = 1
    elif example['personalFemale'] == 1:
        example['gender_male'] = 0
    else:
        example['gender_male'] = None  # Handling cases where neither is 1
    return example

# Apply the transformation to the sampled dataset
sampled_ds = sampled_ds.map(add_gender_column)

# Verify the result
print(sampled_ds.features)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test= train_test_split(sampled_ds['image'], sampled_ds['gender_male'], test_size=0.2, random_state=42)
```

```python
type(X_train[10])
```

```python
import tensorflow as tf
from tensorflow.keras.applications.resnet50 import preprocess_input
import numpy as np
from PIL import Image

# Path to your image dataset (adjust this to your dataset's directory)
IMAGE_SIZE = (224, 224)  # Size required by ResNet or any CNN model

def load_and_preprocess_image(img):
    """
    Preprocess the PIL image: Convert to array, resize it, and normalize it for a model like ResNet.
    """
    if not isinstance(img, Image.Image):
        raise TypeError(f"Expected PIL Image, got {type(img)}")

    img = img.resize(IMAGE_SIZE, Image.Resampling.LANCZOS)
    img_array = np.array(img)

    if img_array.ndim == 2:  # Handle grayscale images
        img_array = np.expand_dims(img_array, axis=-1)
        img_array = np.repeat(img_array, 3, axis=-1)

    img_array = preprocess_input(img_array)

    return img_array

def preprocess_images(images):
    """
    Preprocess a list of images, skipping any that are not PIL images.
    """
    preprocessed_images = []
    for img in images:
        try:
            preprocessed_image = load_and_preprocess_image(img)
            preprocessed_images.append(preprocessed_image)
        except TypeError:
            print(f"Skipping invalid image of type {type(img)}")

    return np.array(preprocessed_images)

# Verify types and shapes before preprocessing
print("Preprocessing training images...")
X_train_preprocessed = preprocess_images(X_train)

print("Preprocessing test images...")
X_test_preprocessed = preprocess_images(X_test)

# Verify the shape of the preprocessed images
print(f"Shape of preprocessed training images: {X_train_preprocessed.shape}")
print(f"Shape of preprocessed test images: {X_test_preprocessed.shape}")
```

```python
import tensorflow as tf
from tensorflow.keras.applications import ResNet50
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D
from tensorflow.keras.optimizers import Adam

# Load the ResNet50 model with pretrained weights and exclude the top layers
base_model = ResNet50(weights='imagenet', include_top=False, input_shape=(224, 224, 3))

# Freeze the base model to avoid modifying the pretrained weights
base_model.trainable = False

# Add custom layers on top of the ResNet base
x = base_model.output
x = GlobalAveragePooling2D()(x)  # Add a global average pooling layer
x = Dense(1024, activation='relu')(x)  # Add a fully connected layer
predictions = Dense(1, activation='sigmoid')(x)  # Output layer for binary classification

# Define the model
model = Model(inputs=base_model.input, outputs=predictions)
```

```python
 #Compile the model
model.compile(optimizer=Adam(learning_rate=1e-4),
              loss='binary_crossentropy',
              metrics=['accuracy'])

# Print model summary
model.summary()
```

```python
print(type(X_train_preprocessed))  # Should be <class 'numpy.ndarray'>
print(X_train_preprocessed.shape)  # Should be (num_samples, height, width, channels)
```

```python
import numpy as np

# Convert lists to NumPy arrays
y_train = np.array(y_train)
y_test = np.array(y_test)

# Print shapes to verify
print(y_train.shape)
print(y_test.shape)
```

```python
# Example: Training the model
# Assuming you have X_train_preprocessed, y_train, X_test_preprocessed, and y_test
history = model.fit(X_train_preprocessed, y_train,
                    validation_data=(X_test_preprocessed, y_test),
                    epochs=50,
                    batch_size=32)

```

```python
# Evaluate the model
test_loss, test_acc = model.evaluate(X_test_preprocessed, y_test)
print(f"Test accuracy: {test_acc}")
```
